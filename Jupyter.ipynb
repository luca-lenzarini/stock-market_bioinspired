```python
from sklearn.metrics import roc_curve, roc_auc_score
import matplotlib.pyplot as plt
import datetime
from main import *

inicio = datetime(2020, 1, 1)
fim = datetime(2020, 1, 10)


test = []
pred = []

for i in range(10):
    a = exec_test("petr4", ["OIBR3", "ITUB4", "VVAR3"], inicio.strftime("%m-%d-%Y"), fim.strftime("%m-%d-%Y"), 'Abertura', 5)

    inicio = inicio + timedelta(days=7)
    fim = fim + timedelta(days=7)

    test += a["result"]
    pred += a["Firefly Algorithm"]

auc = roc_auc_score(test, pred)

fpr, tpr, _ = roc_curve(test, pred)

r_probs = [0 for _ in range(len(test))]
r_auc = roc_auc_score(test, r_probs)

r_fpr, r_tpr, _ = roc_curve(test, r_probs)

plt.plot(r_fpr, r_tpr, linestyle='--', label='Random prediction (AUROC = %0.3f)' % r_auc)
plt.plot(rf_fpr, rf_tpr, marker='.', label='Random Forest (AUROC = %0.3f)' % auc)

# Title
plt.title('ROC Plot')
# Axis labels
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
# Show legend
plt.legend() # 
# Show plot
plt.show()

```
